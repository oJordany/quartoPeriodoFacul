# Método da Eliminação de Gauss com o Pivoteamento Parcial


```python
import pandas as pd
import time

def eliminacaoDeGauss(A, b):
  tempo_inicial = time.time()
  # qntd_iteracoes = 0
  for k in range(1, len(A)):
    # qntd_iteracoes += 1
    # PIVOTEAMENTO PARCIAL
    for i in range(k+1, len(A)+1):
      # qntd_iteracoes += 1
      if (abs(A[i-1][k-1]) > abs(A[k-1][k-1])):
        [A[k-1], A[i-1]] = [A[i-1], A[k-1]]
        [b[k-1], b[i-1]] = [b[i-1], b[k-1]]
    # ELIMINAÇÃO
    for i in range(k+1, len(A)+1):
      # qntd_iteracoes += 1
      if (A[i-1][k-1] != 0):
        m = A[i-1][k-1]/A[k-1][k-1]
        A[i-1][k-1] = 0
        for j in range(k+1, len(A)+1):
          # qntd_iteracoes += 1
          A[i-1][j-1] = A[i-1][j-1] - m * A[k-1][j-1]
        b[i-1] = b[i-1] - m * b[k-1]
  # RESOLUÇÃO DO SISTEMA
  x = [None for _ in range(0, len(b))]
  xn = b[-1]/A[-1][-1]
  x[-1] = xn
  for k in range(len(b) - 2, -1, -1):
    # qntd_iteracoes += 1
    s = 0
    for j in range(k+1, len(b)):
      # qntd_iteracoes += 1
      s = s + A[k][j]*x[j]
    x[k] = (b[k] - s)/A[k][k]

  tempo_final = time.time()
  tempo_de_execucao = tempo_final - tempo_inicial
  return x


A = [[7, 5, 5, 2], [1, 7, 7, 8], [4, 10, 6, 5], [8, 9, 2, 0]]
b = [9,3,-2,1]
Tmatrix = eliminacaoDeGauss(A,b)

print(Tmatrix)
# A = pd.DataFrame(Tmatrix[0], index=[f'linha {i}' for i in range (1, len(Tmatrix[0])+1)])
# A.index.name = "Valores de A"
# A.columns = [f'coluna {i}' for i in range (1, len(Tmatrix[0])+1)]
# b = pd.DataFrame(Tmatrix[1], index=[f'linha {i}' for i in range (1, len(Tmatrix[1])+1)])
# b.index.name = "Valores de b"
# b.columns = ['coluna 1']
# x = pd.DataFrame(Tmatrix[2], index=[f'x{i}' for i in range (1, len(Tmatrix[2])+1)])
# x.index.name = "Valores de x"
# x.columns = ['coluna 1']
# display(A)
# display(b)
# display(x)
```

# Fatoração LU com pivoteamento parcial

```python
import time

def fatoracaoLU(A, b):
  # tempo_inicial = time.time()
  # PIVOTEAMENTO PARCIAL
  P = list(range(1, len(A)+1))
  qntd_iteracoes = 0
  for k in range(1, len(A)):
    qntd_iteracoes += 1
    pv = abs(A[k-1][k-1])
    r = k 
    for i in range(k+1, len(A)+1):
      qntd_iteracoes += 1
      if (abs(A[i-1][k-1]) > pv):
        pv = abs(A[i-1][k-1])
        r = i

    if pv == 0:
      print('A matriz é singular')
      # tempo_final = time.time()
      # tempo_de_execucao = tempo_final - tempo_inicial
      return [False, qntd_iteracoes]
    if r != k:
      aux = P[k-1] 
      P[k-1] = P[r-1]
      P[r-1] = aux
      for j in range(1, len(A)+1):
        qntd_iteracoes += 1
        aux = A[k-1][j-1]
        A[k-1][j-1] = A[r-1][j-1]
        A[r-1][j-1] = aux 

  # ELIMINAÇÃO   
    for i in range(k+1, len(A)+1):
      qntd_iteracoes += 1
      m = A[i-1][k-1]/A[k-1][k-1]
      A[i-1][k-1] = m
      for j in range(k+1, len(A)+1):
        qntd_iteracoes += 1
        A[i-1][j-1] = A[i-1][j-1] - m*A[k-1][j-1]

  # RESOLUÇÃO DOS SISTEMAS 

  # c = Pb
  c = [None for _ in range(0, len(b))]
  for i in range(1, len(A)+1):
    qntd_iteracoes += 1
    r = P[i-1]
    c[i-1] = b[r-1]

  # Ly = c
  y = [c[i]/A[i][i] if i == 0 else None for i in range(0, len(b))]

  for i in range(1, len(A)+1):
    qntd_iteracoes += 1
    soma = 0
    for j in range(1, i):
      qntd_iteracoes += 1
      soma = soma + A[i-1][j-1]*y[j-1]
    y[i-1] = c[i-1] - soma

  # Ux = y
  x = [y[i]/A[i][i] if i == len(b)-1 else None for i in range(0, len(b))]
  
  for i in range(len(A), 0, -1):
    qntd_iteracoes += 1
    soma = 0
    for j in range(i + 1, len(A)+1):
      qntd_iteracoes += 1
      soma = soma + A[i-1][j-1] * x[j-1]
    x[i-1] = (y[i-1] - soma)/A[i-1][i-1]
  # tempo_final = time.time()
  # tempo_de_execucao = tempo_final - tempo_inicial
  return [x, qntd_iteracoes]

A = [[7, 5, 5, 2], [1, 7, 7, 8], [4, 10, 6, 5], [8, 9, 2, 0]]
b = [9,3,-2,1]
solution = fatoracaoLU(A,b)
print(solution)
# A = pd.DataFrame(solution[0], index=[f'linha {i}' for i in range (1, len(solution[0])+1)])
# A.index.name = "Valores de A"
# A.columns = [f'coluna {i}' for i in range (1, len(solution[0])+1)]
# b = pd.DataFrame(solution[1], index=[f'linha {i}' for i in range (1, len(solution[1])+1)])
# b.index.name = "Valores de b"
# b.columns = ['coluna 1']
# y = pd.DataFrame(solution[2], index=[f'linha {i}' for i in range (1, len(solution[2])+1)])
# y.index.name = "Valores de y"
# y.columns = ['coluna 1']
# x = pd.DataFrame(solution[3], index=[f'x{i}' for i in range (1, len(solution[3])+1)])
# x.index.name = "Valores de x"
# x.columns = ['coluna 1']

# display(A)
# display(b)
# display(y)
# display(x)
```

# Método Iterativo de Gauss-Jacobi

```python
import pandas as pd
import math 
import time

def buscadorDePrecisao():
  A = 1
  s = 2

  while s > 1: 
    A = A/2 
    s = 1 + A 

  return A*2

def comparar(x, xk, precisao):
  soma = 0
  distancias = []
  for i in range(0, len(x)):
    distancias.append(x[i] - xk[i])
  
  if max(distancias) < precisao:
    return True
  else:
    return False

def gaussJacobi(A, b, iteracaoMaxima=10, precisao=0.001): 
  tempo_inicial = time.time()
  n = len(b)
  sol = True
  x = b.copy()
  aux = []
  for i in range(1, n+1):
    acc = 0
    for j in range(1, n+1):
      acc += A[i-1][j-1] if i != j else 0 
    if A[i-1][i-1] != 0 :
      aux.append(acc/A[i-1][i-1])
    else:
      tempo_final = time.time()
      tempo_de_execucao = tempo_final - tempo_inicial
      return [False, tempo_de_execucao]

  if max(aux) >= 1:
    print(f"Não converge: {aux.index(max(aux))}")
    tempo_final = time.time()
    tempo_de_execucao = tempo_final - tempo_inicial
    return [False, tempo_de_execucao]

  for i in range(1, n+1):
    if (abs(A[i-1][i-1] != 0)):
      x[i-1] = b[i-1]/A[i-1][i-1]
    else:
      sol = False
      break 
  
  if (sol):
    # print("Iteração 0")
    # print("x = ", x)
    xk = x.copy()
    iter = 0

    while(iter < iteracaoMaxima):
      iter += 1
      for i in range(1, n+1):
        soma = 0
        for j in range(1, n+1):
          if i-1 != j-1:
            soma = soma + A[i-1][j-1] * x[j-1]
        
        xk[i-1] = (1/A[i-1][i-1])*(b[i-1] - soma)
      
      # print("Iteração ", iter)
      # print("xk = ", xk)
      if comparar(x, xk, precisao):
        x = xk.copy()
        break
      x = xk.copy()
  tempo_final = time.time()
  return x

A =  [[1.0000000155844808, 1.44398926121008e-08, 1.4534133468755783e-08, 1.2953942080320021e-08], [1.6345314840497912e-08, 1.0000000153240483, 1.4540703446444607e-08, 1.2736863125284542e-08], [1.3242907508395088e-08, 1.4443897252762467e-08, 1.000000015099755, 1.3326080232626778e-08], [1.3237660909389293e-08, 1.2571732088589691e-08, 1.1580397495819733e-08, 1.000000016154496]]
b =  [0.44199403271906906, 0.5414903492891614, 0.2301449951920902, 0.5377470012031456]
# A = [[10,2,1],
#      [1,5,1],
#      [2,3,10]]
# b = [7,-8,6]
precisao = buscadorDePrecisao()
x = gaussJacobi(A, b, 1000000, precisao)
Xdf = pd.DataFrame(x)
display(Xdf)
```

# Método Iterativo de Gauss-Seidel

```python
import pandas as pd
import math 
import time

def buscadorDePrecisao():
  A = 1
  s = 2

  while s > 1: 
    A = A/2 
    s = 1 + A 

  return A*2

def comparar(x, xk, precisao):
  soma = 0
  distancias = []
  for i in range(0, len(x)):
    distancias.append(x[i] - xk[i])
  
  if max(distancias) < precisao:
    return True
  else:
    return False

def gaussSeidel(A, b, iteracaoMaxima=10, precisao=0.001): 
  tempo_inicial = time.time()
  n = len(b)
  sol = True
  x = b.copy()
  aux = []
  for i in range(1, n+1):
    acc = 0
    for j in range(1, n+1):
      if i != 1 and j < i:
        acc += A[i-1][j-1]*aux[j-1]  
      elif j > i:
        acc += A[i-1][j-1]
    if A[i-1][i-1] != 0 :
      aux.append(acc/A[i-1][i-1])
    else:
      return [False]
  print(aux)
  if max(aux) >= 1:
    print(f"Não converge: {aux.index(max(aux))}")
    tempo_final = time.time()
    tempo_de_execucao = tempo_final - tempo_inicial
    return [False, tempo_de_execucao]

  for i in range(1, n+1):
    if (abs(A[i-1][i-1] != 0)):
      x[i-1] = b[i-1]/A[i-1][i-1]
    else:
      sol = False
      break 
  
  if (sol):
    # print("Iteração 0")
    # print("x = ", x)
    xk = x.copy()
    iter = 0

    while(iter < iteracaoMaxima):
      iter += 1
      for i in range(1, n+1):
        soma = 0
        for j in range(1, n+1):
          if i-1 > j-1:
            soma = soma + A[i-1][j-1] * xk[j-1]
          elif (i-1) < (j-1):
            soma = soma + A[i-1][j-1] * x[j-1]
        
        xk[i-1] = (1/A[i-1][i-1])*(b[i-1] - soma)
      
      # print("Iteração ", iter)
      # print("xk = ", xk)
      if comparar(x, xk, precisao):
        x = xk.copy()
        break
      x = xk.copy()


  tempo_final = time.time()
  tempo_de_execucao = tempo_final - tempo_inicial
  return x

A = [[7, 5, 5, 2], [1, 7, 7, 8], [4, 10, 6, 5], [8, 9, 2, 0]]
b = [9,3,-2,1]
precisao = buscadorDePrecisao()
x = gaussSeidel(A, b, 1000000, precisao)
Xdf = pd.DataFrame(x)
display(Xdf)
```

# Gerando sistemas aleatórios

```python
from random import randint
import numpy as np
import json
import pandas as pd


def mudancaAleatoria(matriz):
    n = matriz.shape[0]
    linha = np.random.randint(n)
    if (np.random.rand() > 0.25):
        matriz[linha, :] += np.random.rand()* 10 ** -10 * matriz[np.random.randint(n), :]
    else:
        matriz[linha, :] -= np.random.rand()* 10 ** -10 * matriz[np.random.randint(n), :]


def gerarSistema(n):
    matriz = np.zeros((n, n+1))
    b = np.random.rand(n)
    matriz[:, n] = b
    matriz[:,:-1] = np.identity(n)
    matriz * np.random.randint(400,600)
    rounds = np.random.randint(8000, 10000)
    while rounds > 0:
        mudancaAleatoria(matriz)
        rounds -= 1

    A = matriz[:, :-1].tolist()
    x = b.tolist()
    newb = [element[0] for element in matriz[:, -1:].tolist()]
    
    return [A, x, newb]


for i in range(100, 2000, 100):
    print(i, "\n")
    linearSystem = gerarSistema(i)
    ARandom = linearSystem[0]
    xRandom = linearSystem[1]
    bRandom = linearSystem[2]
    eliminacaoDeGaussMethod = eliminacaoDeGauss(ARandom, bRandom)
    eliminacaoDeGaussMethodSolution = eliminacaoDeGaussMethod[0]
    eliminacaoDeGaussMethodTime = eliminacaoDeGaussMethod[1]
    print(eliminacaoDeGaussMethodSolution)
    print(xRandom)

    try:
        with open('eliminacaoDeGaussTimes_Sparse.json', 'r', encoding="utf-8") as jsonFile:
            datas = json.load(jsonFile)
            datas.append(eliminacaoDeGaussMethodTime)
        with open('eliminacaoDeGaussTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump(datas, jsonFile)  
    except:
        with open('eliminacaoDeGaussTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump([eliminacaoDeGaussMethodTime], jsonFile)

    fatoracaoLUMethod = fatoracaoLU(ARandom, bRandom)
    fatoracaoLUMethodSolution = fatoracaoLUMethod[0]
    fatoracaoLUMethodTime = fatoracaoLUMethod[1]
    print(fatoracaoLUMethodSolution)
    print(xRandom)

    try:
        with open('fatoracaoLUTimes_Sparse.json', 'r', encoding="utf-8") as jsonFile:
            datas = json.load(jsonFile)
            datas.append(fatoracaoLUMethodTime)
        with open('fatoracaoLUTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump(datas, jsonFile)  
    except:
        with open('fatoracaoLUTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump([fatoracaoLUMethodTime], jsonFile)  

    gaussSeidelMethod = gaussSeidel(ARandom, bRandom)
    gaussSeidelMethodSolution = gaussSeidelMethod[0]
    gaussSeidelMethodTime = gaussSeidelMethod[1]
    print(gaussSeidelMethodSolution)
    print(xRandom)

    try:
        with open('gaussSeidelTimes_Sparse.json', 'r', encoding="utf-8") as jsonFile:
            datas = json.load(jsonFile)
            datas.append(gaussSeidelMethodTime)
        with open('gaussSeidelTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump(datas, jsonFile)  
    except:
        with open('gaussSeidelTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump([gaussSeidelMethodTime], jsonFile) 

    gaussJacobiMethod = gaussJacobi(ARandom, bRandom)
    gaussJacobiMethodSolution = gaussJacobiMethod[0]
    gaussJacobiMethodTime = gaussJacobiMethod[1]
    print(gaussJacobiMethodSolution)
    print(xRandom)

    try:
        with open('gaussJacobiTimes_Sparse.json', 'r', encoding="utf-8") as jsonFile:
            datas = json.load(jsonFile)
            datas.append(gaussJacobiMethodTime)
        with open('gaussJacobiTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump(datas, jsonFile)  
    except:
        with open('gaussJacobiTimes_Sparse.json', 'w', encoding="utf-8") as jsonFile:
            json.dump([gaussJacobiMethodTime], jsonFile) 
```

# Gerando os gráficos

```python
import altair as alt
import pandas as pd
import json

def buildChart(timesFile, iterationsFile):
    with open(iterationsFile, 'r', encoding='utf-8') as IterationJsonFile, \
        open(timesFile, 'r', encoding='utf-8') as TimesJsonFile:
        timesList = json.load(TimesJsonFile)
        iterationsList = json.load(IterationJsonFile)

    timesDict = {
        'Dimensão da matriz' : [matrixDimension for matrixDimension in range(100, 2100, 100)],
        'Tempo de execução': timesList,
        'Quantidade de Iterações': iterationsList,
    }

    timesDf = pd.DataFrame(timesDict)

    base = alt.Chart(timesDf, title="Eliminação de Gauss com Pivoteamento").encode(
        alt.X('Dimensão da matriz'),
    )

    timesLine = base.mark_line(color="blue").encode(
        alt.Y('Tempo de execução', axis=alt.Axis(title='Tempo de execução'))
    )

    iterationsLine = base.mark_line(color="red").encode(
        alt.Y('Quantidade de Iterações', axis=alt.Axis(title='Quantidade de Iterações'))
    )

    layer = alt.layer(
        timesLine, 
        iterationsLine
    ).resolve_scale(
        y='independent'
    ).configure_axisLeft(
        labelColor="blue",
        titleColor="blue"
    ).configure_axisRight(
        labelColor="red",
        titleColor="red"
    )
    return layer

gaussEliminationChart = buildChart('gaussEliminationTimes.json', 'gaussEliminationIterations.json')
fatoracaoLUChart = buildChart('fatoracaoLUTimes.json', 'fatoracaoLUIterations.json')
alt.hconcat(gaussEliminationChart, fatoracaoLUChart)
```
